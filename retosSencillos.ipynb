```python
'''
    Escribe una función que reciba un string y devuelva ese mismo string al revés
'''


def invertir_string(cadena):
    return cadena[::-1]

# Uso de la función
texto = input("Escribe el string: ")
print(invertir_string(texto))
```

```python
'''
    Palíndromo: escribe una función que determine si una palabra o frase es un palíndromo 
    Se lee igual al revés ps.
'''
import math
cadena= "AMÑÑDMA"
cadena=cadena.lower()
#print(s.center(40))
palindrome=True
for i in range(math.ceil(len(cadena)/2)):
    print(cadena[i],cadena[-i-1])
    if cadena[i]!=cadena[-i-1]:
        palindrome=False
        break

#print(math.ceil(5.2))
print("¿Es palíndrome la cadena ",cadena,"?",palindrome)
```

```python
'''
    CONTADOR DE PALABRAS
        Escribe una función que reciba unt exto y devuelva un diccionario con la frecuencia de cada palabra
'''

def contadorPalabras(texto):
    separacion=texto.lower().split(" ")
    #print(separacion)
    diccionario = {}
    for i in separacion:
        try:
            diccionario[i]+=1
        except:
            diccionario[i]=1
    return diccionario
texto = input("Escribe un texto :)")
print(contadorPalabras(texto))

#acabo de ver q existe dic(Conter(words))
'''
from collections import Counter
print(dict(Counter(texto.lower().split())))
'''
```

```python
'''
    FIZZBUZZ
        Imprime los números del 1 al 100, pero:
            si es divisible por 3 imprime fizz
            si es div por 5 buzz
            si por ambos fizzbuzz
'''
for i in range(100):
    if (i+1)%5==0 and (i+1)%3==0:
        print("FizzBuzz")
    elif (i+1)%5==0:
        print("Buzz")
    elif (i+1)%3==0:
        print("Fizz")
    else:
        print(i+1)
```

```python
'''
    ANAGRAMAS: Escribe una función que verifique si dos palabras son anagramas (tienen las mismas letras, distinto orden)
'''

def contador(p1,p2):
    return sorted(p1.lower())==sorted(p2.lower())

palabra1 = input("Escribe tu palabra 1: ")
palabra2 = input("Escribe tu palabra 2: ")

print("¿Son Anagramas?", contador(palabra1,palabra2))
    

```

```python
'''
    Imprimir un tablero de ajedrez
'''
n = 8 #tablero
for i in range(n):
    fila = ''
    for j in range(n):
        if (i+j)%2==0:
            fila += '⬜'
        else:
            fila += '⬛'

    print(fila)
```
